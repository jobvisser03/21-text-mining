```python
import pandas as pd
import datetime as dt
```

```python
d = pd.read_table('./data_2013.txt', names=["all"])
d.head()
```

```python
# Extract date from messages with regex
d['date'] = d['all'].str.extract("(\d{1,2}/\d{2}/\d{2}, \d{2}:\d{2})", expand=True)
d['date'] = d['all'].str.extract("(\d{1,2}/\d{2}/\d{2}, \d{2}:\d{2})", expand=True)

# Parse date as datetime. format: 3/25/13, 14:20
d['date'] = pd.to_datetime(d['date'], format='%m/%d/%y, %H:%M')
```

```python
d['name'] = d['all'].str.extract("(\d{1,2}/\d{2}/\d{2}, \d{2}:\d{2})( -.*?:)(.*)", expand=True)[1].str[3:-1]
d['message'] = d['all'].str.extract("(\d{1,2}/\d{2}/\d{2}, \d{2}:\d{2})( -.*?:)(.*)", expand=True)[2]
d.drop('all', 1)
```
